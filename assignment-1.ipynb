# Assignment 1: Temperature Check
#### Problem:
  - Write a program that takes the current temperature as input and prints
    - "It's too cold!" if below 10°C
    - "It's cool outside" if between 10°C and 25°C
    - "It's hot outside!" if above 25°C

```python
temp = int(input("Enter the currnt temperature"))
if temp <10:
    print("It's too cold!")
elif temp >10 and temp <25:
    print("It's cool outside.")
else:
    print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System
#### Problem:
- Ask the user for their account balance and the amount they want to withdraw.
    - If the withdrawal amount is greater than the balance → print "Insufficient
Balance"
    - If the withdrawal amount is less than or equal to balance → print "Transaction
Successful" and show the remaining balance.

```python
balance = float(input("Enter your account balance"))
print(f"My balance is {balance} RS")
amount = float(input("Enter your withdraw amount"))
print(f"My withdraw amount is {amount} RS")
if amount > balance:
    print("Insufficient Balance")
else:
    if amount <= balance:
        print(f"Transaction Successful and remaining balance is {balance - amount} RS")
```

# Assignment 3: Student Grade Calculator
#### Problem:
- Input marks (0–100) and print grade based on:
  - 90–100 → A+
  - 80–89 → A
  - 70–79 → B
  - 60–69 → C
  - Below 60 → Fail

```python
marks = int(input("Enter your marks between 0 to 100."))
if marks >= 90 and marks <=100:
    print(f"Your marks is {marks} and Your Grade: A+")
elif marks >=80 and marks <=89:
    print(f"Your marks is {marks} and Your Grade: A")
elif marks >=70 and marks <=79:
    print(f"Your marks is {marks} and Your Grade: B")
elif marks >=60 and marks <=69:
    print(f"Your marks is {marks} and Your Grade: C")
else:
    print(f"Your Marks is below {marks} then You are Fail")
```

# Assignment 4: Bus Ticket Discount
#### Problem:
- Ask the user for age and print ticket price:
  - Below 5 → Free
  - 5–18 → 50% Discount
  - 60 or above → 30% Discount
  - Others → Full Price

```python
age = int(input("Enter your Age:"))
print(f"Your age is  {age} years")
ticket_price = int(input("Enter the Bus ticket price"))
print(f"Bus Ticket price {ticket_price} RS")
if age < 5:
    print(f"Your age {age} years You don't buy ticket.Your ride is free!")
elif age >5 and age <=18 :
    ticket_price = ticket_price * 0.5
    print(f"Your age {age} years and your ticket price is {ticket_price} RS")
elif age >= 60:
    ticket_price = ticket_price * 0.3
    print(f"Your age {age} years and your ticket price is {ticket_price} RS")
else:
    print(f"You will have to pay  {ticket_price} RS ")
```

# Assignment 8: Leap Year Checker
#### Problem:
- Take a year as input and check:
  - If it is divisible by 400 → Leap Year
  - Else if divisible by 4 but not by 100 → Leap Year
  - Otherwise → Not a Leap Year

```python
year = int(input("Enter the Year"))
print(f"You select year is {year} year")
if year % 400 == 0:
    print(f"This {year} year Leap Year")
elif (year % 4 == 0) and (year % 100 !=0):
    print(f"This {year} year Leap Year")
else :
    print(f"This {year} year is not Leap Year")
```

# Assignment 10: Electricity Bill Calculator
#### Problem:
- Input number of units consumed and calculate bill:
  - Up to 100 units → 5 BDT per unit
  - 101–200 units → 7 BDT per unit
  - Above 200 units → 10 BDT per unit
  - Print total payable amount.

```python
units = int(input("Enter your units")) 
print(f"Your units is {units}")
if units <=100 :
    bill = 5 * units
    print(f"Your units is {units} and your bill is {bill} rs")
elif units >=101 and units <= 200 :
    bill = 7 * units
    print(f"Your units is {units} and your bill is {bill} rs")
else :
    if units > 200:
        bill = units * 10
        print(f"Your units is {units} and your bill is {bill} rs")
```
