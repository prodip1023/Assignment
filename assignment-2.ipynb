# 1. Email Validator (String Filtering + Loops)
- Real-world use: Used in signup forms or data cleaning tasks.
- Task:
    - Check if a given email is valid — must contain “@” and end with “.com” or “.org”.
- Example:
    - Input: hello@inceptionbd.com
    - Output: Valid Email

```python
email_address = input("Enter your correct email address").strip()
print(f"My Email Address is :{email_address}")
if '@' in email_address and (email_address.endswith('.com') or email_address.endswith('.org')):
    print("This is a valid Email Address")
else:
    print("This is not a valid Email Address")
   
```

# 2. Count Word Frequency in Customer Reviews
- Real-world use: Analyzing reviews or comments in text analytics
- Task:
    - Given a review text, count how many times each word appears (ignore case).
- Example:
    - Input: "This product is good and this service is excellent"
    - Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
user_input = input("Enter the Sentence.").strip()
print(f"Sentence is : {user_input}")
words = user_input.split()
count_words = {word: words.count(word) for word in words}
print(count_words)    
```

# 3. Password Strength Checker
- Real-world use: Used in user authentication systems.
- Task:
    -   Check if a password is strong — must have at least one uppercase, one lowercase, one digit, and one
special character.
- Example:
    -   Input: "Hello@123"
    -   Output: Strong Password

```python
password = input("Enter the password,atleast one uppercase,one lowercase,one digit and one special character").strip()
print(f"This is my password: {password}")
upper_case = False
lower_case = False
digit = False
special_case = False

for pawd in password:
    if pawd.isupper():
        upper_case = True
    elif pawd.islower():
        lower_case = True
    elif pawd.isdigit():
        digit = True
    else:
        special_case = True

if upper_case and lower_case and digit and special_case:
    print("It's Strong Password")
else:
    print("Weak Password")
```

# 4. Username Generator
- Real-world use: Automatically create usernames for users.
- Task:
    - Take user’s full name and generate a short username.
    - Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).

- Example:
    - Input: "Mamun Malitha"
    - Output: mamitha57


```python
import random
user_name = input("Enter Your User Name").strip()
name = user_name.split()
first_name = name[0]
last_name = name[-1]

fname=""
count = 0
for fn in first_name:
    if count == 3:
        break
    fname +=fn.lower()
    count+=1

lname = ""
count = 0
for ln in reversed(last_name):
    if count == 3:
        break
    lname +=ln.lower()
    count+=1

num = random.randint(10, 99)
username = fname + lname + str(num)

print("Generated Username:", username)


   
```

# 5. SMS Word Counter
- Real-world use: Used in messaging apps or social media post limit check.
- Task:
    - Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).
- Example:
    - Input: "This is a demo message."
    - Output: Words = 5, Characters = 23, Status = Within Limit

```python
message = input("Enter the message").strip()
words = len(message.split())
characters = len(message)
if characters > 160:
    status = "Exceeds Limit"
else:
    status = "Within Limit"

print(f"Words = {words}, Characters = {characters}, Status = {status}")
```

# 6. Remove Duplicates from Contact List
- Real-world use: Data cleaning in CRM or user databases.
- Task:
    - Given a list of names (comma separated), remove duplicates.
- Example:
    - Input: "Rahim, Karim, Rahim, Sultana"
    - Output: "Rahim, Karim, Sultana"

```python
names = ["Rahim","Karim","Rahim","Sultana"]
print(f"List of names :{names}")
unique_names = []
for name in names:
    if name not in unique_names:
        unique_names.append(name)

print(f"Unique names : {unique_names}")
```

# 7. Detect Suspicious Words in Text (Moderation)
- Real-world use: Used in chat moderation or email filtering.
- Task:
    - If any “banned” words like ['spam', 'scam', 'fake'] are found in a text, print “Warning”.
- Example:
    - Input: "This is a scam offer"
    - Output: Warning: Suspicious content detected
  

```python
text = input("Enter your text: ").lower()
banned_words = ['spam', 'scam', 'fake']
warning = False
for word in banned_words:
    if word in text:
        warning = True
        break
if warning:
    print("Warning: Suspicious content detected")
else:
    print("Text is clean")
```

# 10. Word Reversal for Data Encryption
- Real-world use: Simple text obfuscation or playful encoding
- Task:
    - Reverse each word in a sentence while keeping word order the same.
- Example:
    - Input: "AI is fun"
    - Output: "IA si nuf

```python
sentence = input("Enter your sentence").strip()
words = sentence.split()
reversed_words = []
for word in words:
    reversed_words.append(word[::-1])
reversed_sentence = " ".join(reversed_words)
print("Reversed Sentence:", reversed_sentence)
```
